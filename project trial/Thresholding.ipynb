```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('00001.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap='gray')
```

```python
ret,th1 =  cv2.threshold(img, int(img.max()/2), 255 , cv2.THRESH_BINARY)
plt.imshow(th1,cmap='gray')
```

```python
th2 = cv2.adaptiveThreshold(img , 255 , cv2.ADAPTIVE_THRESH_MEAN_C , cv2.THRESH_BINARY , 9 ,8)
plt.imshow(th2,cmap='gray')
```

```python
final_th = cv2.addWeighted(th1, 0.8 , th2, 0.2, 0)
plt.imshow(final_th,cmap='gray')
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


img = cv2.imread('00001.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
# plt.imshow(img, cmap='gray')


ret,th1 =  cv2.threshold(img, int(img.max()/2), 255 , cv2.THRESH_BINARY)
# plt.imshow(th1,cmap='gray')


th2 = cv2.adaptiveThreshold(img , 255 , cv2.ADAPTIVE_THRESH_MEAN_C , cv2.THRESH_BINARY , 9 ,8)
# plt.imshow(th2,cmap='gray')

final_th = cv2.addWeighted(th1, 0.8 , th2, 0.2, 0)
plt.imshow(final_th,cmap='gray')
```
