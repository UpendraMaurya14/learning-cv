```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def display(img,cmap=None):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
img = cv2.cvtColor(cv2.imread('balconyTile.jpeg'),cv2.COLOR_BGR2RGB)
gray_img = cv2.cvtColor(cv2.imread('balconyTile.jpeg'),cv2.COLOR_BGR2GRAY)
```

```python
blur_img = cv2.blur(img, (9,9))
# plt.imshow(img)
display(blur_img)
```

```python
blur_img = cv2.blur(gray_img, (3,3))
# plt.imshow(img)
# display(blur_img)
med = np.median(blur_img)
lower = int(max(0,0.7*med))
upper = int(min(255,1.3*med))
edges = cv2.Canny(blur_img,lower,upper)
display(edges,'gray')
```
