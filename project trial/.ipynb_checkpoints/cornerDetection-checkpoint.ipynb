```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def display(img,cmap=None):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
img = cv2.cvtColor(cv2.imread('00001.jpg'),cv2.COLOR_BGR2RGB)
gray_img = cv2.cvtColor(cv2.imread('00001.jpg'),cv2.COLOR_BGR2GRAY)
```

```python
plt.subplot(121)
plt.title('color image')
plt.imshow(img)
plt.subplot(122)
plt.title('grayscale image')
plt.imshow(gray_img,cmap='gray')
plt.suptitle('Both the images')
```

```python
gray_img = np.float32(gray_img)
dst = cv2.cornerHarris(gray_img,2,3,0.04)
dst = cv2.dilate(dst,None)
# plt.imshow(dst,cmap = 'gray')
img[dst>0.01*dst.max()] = [255,0,0]
plt.imshow(img)
```

```python
bruno_img = cv2.cvtColor(cv2.imread('bruno.jpeg'),cv2.COLOR_BGR2RGB)
gray_bruno_img = cv2.cvtColor(cv2.imread('bruno.jpeg'),cv2.COLOR_BGR2GRAY)
```

```python
gray_bruno_img = np.float32(gray_bruno_img)
dst = cv2.cornerHarris(gray_bruno_img,2,3,0.04)
dst = cv2.dilate(dst,None)
bruno_img[dst>0.01*dst.max()] = [255,0,0]
# plt.imshow(bruno_img)
fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(111)
ax.imshow(bruno_img)
```
