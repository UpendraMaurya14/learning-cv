```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.cvtColor(cv2.imread('internal_external.png'),cv2.COLOR_BGR2GRAY)
```

```python
img.shape
```

```python
plt.imshow(img,cmap = 'gray')
```

```python
image,contour,hierarchy = cv2.findContours(img,cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)
```

```python
# hierarchy
```

```python
ext_contours = np.zeros(image.shape)
```

```python
for i in range(len(contour)):
    if(hierarchy[0][i][3] == -1):
        cv2.drawContours(ext_contours,contour,i,255,-1)
```

```python
plt.imshow(ext_contours,cmap = 'gray')
```

```python
image,contours,hierarchy = cv2.findContours(img,cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)

interal_contours = np.zeros(img.shape)

for i in range(len(contours)):
    if(hierarchy[0][i][3]!=0-1):
        cv2.drawContours(interal_contours,contours,i,255,-1)

plt.imshow(interal_contours,cmap = 'gray')

```
