```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flat_chessboard = cv2.cvtColor(cv2.imread('flat_chessboard.png'),cv2.COLOR_BGR2RGB)
gray_flat_chessboard = cv2.cvtColor(cv2.imread('flat_chessboard.png'),cv2.COLOR_BGR2GRAY)
real_chessboard = cv2.cvtColor(cv2.imread('real_chessboard.jpg'),cv2.COLOR_BGR2RGB)
gray_real_chessboard = cv2.cvtColor(cv2.imread('real_chessboard.jpg'),cv2.COLOR_BGR2GRAY)
```

```python
corner = cv2.goodFeaturesToTrack(gray_flat_chessboard,100,0.01,10)

corner = np.int0(corner)

for i in corner:
    x,y = i.ravel()
    
    cv2.circle(flat_chessboard,(x,y),5,(255,0,0),-1)
    
plt.imshow(flat_chessboard)
```

```python
real_chessboard = cv2.cvtColor(cv2.imread('real_chessboard.jpg'),cv2.COLOR_BGR2RGB)
gray_real_chessboard = cv2.cvtColor(cv2.imread('real_chessboard.jpg'),cv2.COLOR_BGR2GRAY)

corner = cv2.goodFeaturesToTrack(gray_real_chessboard,100,0.01,10)

corner = np.int0(corner)

for i in corner:
    x,y = i.ravel()
    cv2.circle(real_chessboard,(x,y),5,(255,0,0),-1)
    
plt.imshow(real_chessboard)
```
