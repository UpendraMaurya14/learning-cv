```python
face_cascade = cv2.CascadeClassifier('Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/haarcascades/haarcascade_frontalface_default.xml')
```

```python
def face(img):
    face_img=img.copy()
    face_dimensions = face_cascade.detectMultiScale(face_img)
    for x,y,w,h in face_dimensions:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
    return face_img 


def detect_face(img):
    
  
    face_img = img.copy()
  
    face_rects = face_cascade.detectMultiScale(face_img) 
    
    for (x,y,w,h) in face_rects: 
        cv2.rectangle(face_img, (x,y), (x+w,y+h), (255,255,255), 10) 
        
    return face_img
    
```
