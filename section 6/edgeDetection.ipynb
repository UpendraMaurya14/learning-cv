```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dog_face = cv2.imread('sammy_face.jpg')
dog_face = cv2.cvtColor(dog_face,cv2.COLOR_BGR2RGB)
plt.imshow(dog_face)
```

```python
edges = cv2.Canny(dog_face,127,127)
```

```python
plt.imshow(edges)
```

```python
med = np.median(dog_face)
```

```python
lower = int(max(0,0.7*med))
upper = int(min(255,1.3*med))
```

```python
edges = cv2.Canny(dog_face,lower,upper)
plt.imshow(edges)
```

```python
blur_img = cv2.blur(dog_face,(5,5))
```

```python
edges = cv2.Canny(blur_img,lower,upper+50)
plt.imshow(edges)
```
