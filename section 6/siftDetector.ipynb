```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img,cmap='gray'):
    fig=plt.figure(figsize=(12,10))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
resses = cv2.imread('reeses_puffs.png',0)
cereal = cv2.imread('many_cereals.jpg',0)
```

```python
display(cereal)
```

```python
sift = cv2.xfeatures2d.SIFT_create()
```

```python
kp1,des1 = sift.detectAndCompute(resses,None)
kp2,des2 = sift.detectAndCompute(cereal,None)
```

```python
bf = cv2.BFMatcher()
```

```python
match = bf.knnMatch(des1,des2,k=2)
```

```python
good = []
for i,j in match:
    if i.distance < 0.30*j.distance:
        good.append([i])
```

```python
len(good)
```

```python
len(match)
```

```python
sift_matches = cv2.drawMatchesKnn(resses,kp1,cereal,kp2,good,None,flags=2)
```

```python
display(sift_matches)
```
