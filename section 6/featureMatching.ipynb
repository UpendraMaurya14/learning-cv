```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img,cmap='gray'):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
    
```

```python
reeses = cv2.imread("reeses_puffs.png",0)
cereal = cv2.imread('many_cereals.jpg',0)
```

```python
orb = cv2.ORB_create()
```

```python
kp1,des1=orb.detectAndCompute(reeses,None)
kp2,des2=orb.detectAndCompute(cereal,None)
```

```python
bf = cv2.BFMatcher(cv2.NORM_HAMMING,crossCheck = True)
```

```python
matches = bf.match(des1,des2)
```

```python
matches = sorted(matches,key=lambda x:x.distance)
```

```python
match_reeses = cv2.drawMatches(reeses,kp1,cereal,kp2,matches[:25],None,flags=2)
```

```python
display(match_reeses)
```
