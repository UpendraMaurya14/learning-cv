```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_img():
    img = cv2.cvtColor(cv2.imread('bricks.jpg'),cv2.COLOR_BGR2RGB).astype(np.float32)/255.0
    return img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```python
img = load_img()
display_img(img)
```

```python
res = np.power(img , 1)
display_img(res)
```

```python
img = load_img()
fonts = cv2.FONT_HERSHEY_COMPLEX
img = cv2.putText(img , 'Bricks' , (10,600) , fonts , 8 , (255,0,0) , 4 )
display_img(img)
```
