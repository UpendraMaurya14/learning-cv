```python
img = cv2.imread('crossword.jpg',0)
```

```python
def show_pic(img):
    fig = plt.figure(figsize = (15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
show_pic(img)
```

```python
ret , th1 = cv2.threshold(img , 165 , 255 , cv2.THRESH_BINARY)
show_pic(th1)
```

```python
th2 = cv2.adaptiveThreshold(img , 255 , cv2.ADAPTIVE_THRESH_MEAN_C , cv2.THRESH_BINARY , 7 , 8)
show_pic(th2)
```

```python
final_th = cv2.addWeighted(th1, 0.8 , th2, 0.2,0)
show_pic(final_th)
```
