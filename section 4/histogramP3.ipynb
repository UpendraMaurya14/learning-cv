```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
gorilla = cv2.imread('gorilla.jpg',0)
plt.imshow(gorilla,cmap = 'gray')
```

```python
def display(img,cmap1=None):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap = cmap1)
```

```python
display(gorilla,'gray')
```

```python
hist = cv2.calcHist([gorilla],[0],None,[256],[0,256])
plt.plot(hist)
```

```python
eq_gorilla = cv2.equalizeHist(gorilla)
display(eq_gorilla,'gray')
```

```python
hist2 = cv2.calcHist([eq_gorilla],[0],None,[256],[0,256])
plt.plot(hist2)
```

```python
color_gorilla = cv2.imread('gorilla.jpg')
show_color_gorilla = cv2.cvtColor(color_gorilla,cv2.COLOR_BGR2RGB)
display(show_color_gorilla)
```

```python
hsv_gorilla = cv2.cvtColor(color_gorilla,cv2.COLOR_BGR2HSV)
hsv_gorilla[:,:,2] = cv2.equalizeHist(hsv_gorilla[:,:,2])
show_color_gorilla_eq = cv2.cvtColor(hsv_gorilla,cv2.COLOR_HSV2RGB)
display(show_color_gorilla_eq)
```

```python
img = color_gorilla
color = ('b','g','r')
for i, col in enumerate(color):
    hist = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(hist,color = col)
plt.title('histogram for gorilla image')
plt.show()
```
