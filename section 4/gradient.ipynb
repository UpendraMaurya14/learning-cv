```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img = cv2.imread('sudoku.jpg',0)
display(img)
```

```python
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,5)
display(sobelx)
```

```python
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,5)
display(sobely)
```

```python
blended = cv2.addWeighted(sobelx,0.5,sobely,0.5,0)
display(blended)
```

```python
# laplace = cv2.Laplacian(img,cv2.CV_64F)
# display(laplace)
```

```python
ret,th1 = cv2.threshold(img,img.max()/2,img.max(),cv2.THRESH_BINARY)
display(th1)
```

```python
ret, th2 = cv2.threshold(blended,blended.max()/2,blended.max(),cv2.THRESH_BINARY)
display(th2)
```

```python
kernel = np.ones((5,5))
morph = cv2.morphologyEx(blended,cv2.MORPH_GRADIENT,kernel)
display(morph)
```
