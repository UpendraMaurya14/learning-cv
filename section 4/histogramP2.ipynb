```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rainbow = cv2.imread('rainbow.jpg')
show_rainbow = cv2.cvtColor(rainbow,cv2.COLOR_BGR2RGB)
```

```python
img =rainbow
```

```python
mask = np.zeros(img.shape[:2],np.uint8)
plt.imshow(mask,cmap = 'gray')
```

```python
# mask[200:300,0:400] = 255
# plt.imshow(mask,cmap = 'gray')
```

```python
mask[300:400,100:400] = 255
plt.imshow(mask,cmap = 'gray')
```

```python
mask_rainbow = cv2.bitwise_and(img,img,mask=mask)
mask_show_rainbow = cv2.bitwise_and(show_rainbow,show_rainbow,mask=mask)
plt.imshow(mask_show_rainbow)
```

```python
img  = rainbow
color = ('b','g','r')
for i,col in enumerate(color):
    hist = cv2.calcHist([img],[i],mask,[256],[0,256])
    plt.plot(hist,color = col)
plt.title('histogram for masked region')
plt.show()
```

```python
img = rainbow
color = ('b','g','r')
for i,col in enumerate(color):
    hist = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(hist,color=col)
plt.title('histogram for rainbow')
plt.show()
```
