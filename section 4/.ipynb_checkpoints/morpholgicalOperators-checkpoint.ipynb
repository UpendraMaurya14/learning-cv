```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def loadImage():
    img = np.zeros((600,600))
    cv2.putText(img,'ABCDE',(50,300), cv2.FONT_HERSHEY_SIMPLEX, 5,(255,255,255),25,lineType=cv2.LINE_AA)
    return img;
def displayImg(img):
    fig = plt.figure(figsize = (12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap = 'gray')
```

```python
img = loadImage()
displayImg(img)
```

```python
img = loadImage()
kernel = np.ones((5,5),np.uint8)
erosion1 = cv2.erode(img,kernel,iterations = 4)
displayImg(erosion1)
```

```python
img = loadImage()
kernel = np.ones((5,5),np.uint8)
erosion5 = cv2.erode(img,kernel,iterations = 4)
```

```python
displayImg(erosion5)
```

```python
img = loadImage()
noice_img = np.random.randint(low=0,high=2,size=(600,600))
displayImg(noice_img)
```

```python
noice_img = noice_img * 255;
white_noice_img = noice_img + img
displayImg(white_noice_img)
```

```python
opening = cv2.morphologyEx(white_noice_img,cv2.MORPH_OPEN,kernel)
displayImg(opening)
```

```python
black_noice = np.random.randint(0,2,(600,600))*-255
black_noice_img = black_noice + img
black_noice_img[black_noice_img==-255] = 0
displayImg(black_noice_img)
```

```python
closing = cv2.morphologyEx(black_noice_img,cv2.MORPH_CLOSE,kernel)
displayImg(closing)
```

```python
gradient = cv2.morphologyEx(img , cv2.MORPH_GRADIENT , kernel)
displayImg(gradient)
```
